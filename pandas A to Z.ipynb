```python
#merge
import pandas as pd
var1=pd.DataFrame({"names":["jon","anna","linda","peter"],"age":[28,24,25,33]})#left
var2=pd.DataFrame({"names":["jon","anna","lucy","tom"],"country":["usa","uk","canada","austerlia"]})#RIGHT
pd.merge(var1,var2,on="names",how="left")
pd.merge(var1,var2,on="names",how="right")
pd.merge(var1,var2,on="names",how="outer",)
pd.merge(var1,var2,on="names",how="outer",indicator=True)#indicator creates own new option  names -merge 
```

```python
import pandas as pd
var1=pd.DataFrame({"names":["jon","anna","linda","peter"],"age":[28,24,25,33]})#left
var2=pd.DataFrame({"names":["jon","anna","lucy","tom"],"country":["usa","uk","canada","austerlia"]})#RIGHT
pd.merge(var1,var2,left_index= True, right_index=True,)
pd.merge(var1,var2,left_index= True, right_index=True,suffixes=("name","python"))
```
